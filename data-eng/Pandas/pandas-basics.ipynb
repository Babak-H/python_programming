What differentiates the NumPy array from a Series, is that a Series can have axis labels, meaning it can be indexed by a label, instead of just a number location. It also doesn't need to hold numeric data, it can hold any arbitrary Python Object.

```python
import numpy as np
import pandas as pd
```

### Merging, Joining, and Concatenating

### Merging
The merge function allows you to merge DataFrames together using a similar logic as merging SQL Tables together. Merging needs to happen on specific column name that has same value for both Dataframes.

For example:

```python
left = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

right = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
```

```python
# here we get keys from second DataFrame
pd.merge(left, right, how='inner', on='key')
```

```python
# a more complicated example:
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
# only shows rows that in them both dataframes have same key1, key2
# we have two lines two K1,K0 since one'k1,k0' on left dataframe matches two on right dataframe
pd.merge(left, right, on=['key1', 'key2'])
```

```python
# outter merge, includes NaN value for rows that have no value for that specific key
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
# left merge, all rows from left df are kept, only common rows from right df are kept
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

```python
# right merge, all rows from right df are kept, only common rows from left df are kept
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
# basic example of dataset cleaning
df = pd.read_excel('AFI_market_all4.xlsx')

# Accesses the column named "%Key" in the DataFrame
# applies a lambda function to each value (x) in the "%Key" column
# The lambda function takes the last 7 characters from each string using slicing x[-7:]
# ID2023-06 => 023-06
df['Date'] = df["%Key"].apply(lambda x: x[-7:])

# get characters from 0 to 3
df['SCC'] = df["%Key"].apply(lambda x: x[:4])
```

```python
date_format = '%Y0%m'

df["Date_excel"] = df["Date"].apply(lambda x: datetime.strptime(x, date_format))

# This applies a function to each pair of values from the two columns NetDueDate_PIA and ClearDate_PIA
# If ClearDate_PIA (y) is not "-", it calculates the difference in days between NetDueDate_PIA (x) and ClearDate_PIA (y)
# If y is "-" (which seems to mean it hasnt been cleared yet), it returns "Not cleared"
# [*map(...)] => This expands the map object into a list
df["Duration"] = [
    *map(
        lambda x, y: (x - y).days if y != "-" else "Not cleared",
        df["NetDueDate_PIA"],
        df["ClearDate_PIA"]
    )
]
```

```python
# df["Duration"].apply(...): Applies a function to each value in the "Duration" column
# "Yes": If the duration is a number (either int or float) and it's less than 0, it's considered overdue
# "-": If the value is "Not cleared" (i.e., no clear date yet), mark it with a dash
# "No": Anything else (including 0 or positive duration) is not overdue
df["Overdue_Y_N"] = df["Duration"].apply(
    lambda x: 
        "Yes" if (isinstance(x, (int, float)) and x < 0) 
        else ("-" if x == "Not cleared" 
        else "No")
)
```


JSON isn't a tabular format, so pandas makes assumptions about its orientation when loading data. Most JSON data you encounter will be in orientations that pandas 
can automatically transform into a data frame.
Sometimes, data is oriented differently. To reduce the file size, it has been split formatted. The try/except block will alert you if there are errors loading the data.

```python
try:
    # Load the JSON with orient specified
    df = pd.read_json("dhs_report_reformatted.json", orient="split")
    df["date_of_census"] = pd.to_datetime(df["date_of_census"])
except ValueError:
    print("pandas could not parse the JSON.")

'''
This tells pandas how the JSON file is structured. When orient="split", pandas expects the JSON to be in the following format:
{
  "columns": ["col1", "col2", "col3"],
  "index": [0, 1, 2],
  "data": [
    [val1, val2, val3],
    [val4, val5, val6],
    ...
  ]
}

This format is optimized for performance and compactness when working between Python and pandas

orient="split" just tells pandas what structure to expect inside the JSON file. If the structure doesn't match what split expects, you'll get a ValueError
'''
```
