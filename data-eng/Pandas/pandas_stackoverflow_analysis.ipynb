```python
 # add dataframes side by side
python_df.rename(columns={'count': 'NumRespondants', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
python_df['pctPython'] = (python_df['NumKnowsPython'] / python_df['NumRespondants']) * 100

python_df.head(3)
```

```python
# Dealing with Missing Data
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}

print(type(np.nan), type(None), type('NA'))

df_1 = pd.DataFrame(people)
df_1
```

```python
df_1.replace('NA', np.nan, inplace=True)  # applies to all cells in each row
df_1.replace('Missing', np.nan, inplace=True)
df_1.dropna() # dropna() deletes any line that includes a NaN (np.nan) or None value
```

```python
# only drop rows if email AND last name are missing (both missing)
df_1.dropna(axis='index', how='all', subset=['email', 'last'])
```

```python
df_1.isna() # returns True if value is considered na
```

```python
df_1.fillna('Missing')
```
