```python
'''
You own a Goal Parser that can interpret a string command. The command consists of an alphabet of "G", "()" and/or "(al)" in some order. The Goal Parser will interpret "G" as the string "G", "()" as the string "o", and "(al)" as the string "al". The interpreted strings are then concatenated in the original order.
Given the string command, return the Goal Parser's interpretation of command.

Input: command = "G()(al)"
Output: "Goal"
Explanation: The Goal Parser interprets the command as follows:
G -> G
() -> o
(al) -> al
The final concatenated result is "Goal".

Input: command = "G()()()()(al)"
Output: "Gooooal"

Input: command = "(al)G(al)()()G"
Output: "alGalooG"
'''

class Solution:
    def solve(self, command: str) -> str:
        return command.replace("()", "o").replace("(al)", "al")
    
s = Solution()
print(s.solve("(al)G(al)()()G"))
print(s.solve("G()()()()(al)"))
```

```python
'''
You are given the array paths, where paths[i] = [cityAi, cityBi] means there exists a direct path going from cityAi to cityBi. Return the destination city, that is, the city without any path outgoing to another city.
It is guaranteed that the graph of paths forms a line without any loop, therefore, there will be exactly one destination city.

Input: paths = [["London","New York"],["New York","Lima"],["Lima","Sao Paulo"]]
Output: "Sao Paulo" 
Explanation: Starting at "London" city you will reach "Sao Paulo" city which is the destination city. Your trip consist of: "London" -> "New York" -> "Lima" -> "Sao Paulo".

Input: paths = [["B","C"],["D","B"],["C","A"]]
Output: "A"
Explanation: All possible trips are: 
"D" -> "B" -> "C" -> "A". 
"B" -> "C" -> "A". 
"C" -> "A". 
"A". 
Clearly the destination city is "A".

Input: paths = [["A","Z"]]
Output: "Z"
'''

class Solution:
    def destCity(self, paths: list[list[str]]) -> str:
        outgoing_count = {}
        for path in paths:
            city_a, city_b = path  
            # if the val for the key city_a does not exist, set it to 0, and increase it by one
            outgoing_count[city_a] = outgoing_count.get(city_a, 0) + 1
            # if city is destination, keep its current count, or set it to 0 if it doesnt exist
            outgoing_count[city_b] = outgoing_count.get(city_b, 0)

        for city in outgoing_count:
            if outgoing_count[city] == 0:
                return city
        return ""
    
a = Solution()
a.destCity([["B","C"],["D","B"],["C","A"]])
```

```python
'''
Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. That is, for each nums[i] you have to count the number of valid j's such that j != i and nums[j] < nums[i].
Return the answer in an array.

Input: nums = [8,1,2,2,3]
Output: [4,0,1,1,3]
Explanation: 
For nums[0]=8 there exist four smaller numbers than it (1, 2, 2 and 3). 
For nums[1]=1 does not exist any smaller number than it.
For nums[2]=2 there exist one smaller number than it (1). 
For nums[3]=2 there exist one smaller number than it (1). 
For nums[4]=3 there exist three smaller numbers than it (1, 2 and 2).

Input: nums = [6,5,4,8]
Output: [2,1,0,3]

Input: nums = [7,7,7,7]
Output: [0,0,0,0]
'''

class Solution:
    def smallerNumberThanCurrent(self, nums: list[int]) -> list[int]:
        output = []
        for i in nums:
            count = 0
            for j in nums:
                if j != i and j < i:
                    count += 1
            output.append(count)
        return output
    
s = Solution()
print(s.smallerNumberThanCurrent([8,1,2,2,3]))
print(s.smallerNumberThanCurrent([6,5,4,8]))
print(s.smallerNumberThanCurrent([7,7,7,7]))
```

```python
l = [-1, 0, 3, 5, 9, 12]

# python binary search, we assume that the given list is Sorted
def binary_search(array: list[int], target: int) -> int:
    lower = 0
    upper = len(array)
    while lower < upper:
        mid = lower + (upper - lower) // 2  # middle position of array
        val = array[mid]
        if target == val:
            return mid
        elif target > val:
            if lower == mid:  # we have only 2 elements in remaining list 
                break
            lower = mid  # move the starting line to middle of array
        elif target < val:
            upper = mid  # move ending line to middle of array
    # did Not find the target
    return -1

print(binary_search(l, 9))
print(binary_search(l, 2))
```

```python
'''
Given a string s which consists of lowercase or uppercase letters, return the length of the longest palindrome that can be built with those letters.
Letters are case sensitive, for example, "Aa" is not considered a palindrome.

Input: s = "abccccdd"
Output: 7
Explanation: One longest palindrome that can be built is "dccaccd", whose length is 7.

Input: s = "a"
Output: 1
Explanation: The longest palindrome that can be built is "a", whose length is 1.
'''
# Palindrome: how many characters are repeated evenly, plus one character can be in the middle : xxxZyyy or  ccBaa,...
class Solution:
    def longestPalindrome(self, s: str) -> int:
        char_count = {}
        for char in s:
            char_count[char] = char_count.get(char, 0) + 1

        final_count = 0
        odd_present = False
        for char in char_count:
            if char_count[char] % 2 == 0:
                final_count += char_count[char]
            else:
                odd_present = True
        if odd_present:
            final_count += 1
        return final_count
    
s = Solution()
print(s.longestPalindrome("abccccdd"))
print(s.longestPalindrome("aa"))
```

```python
# Write a script to find all files larger than 100MB in a directory tree.
import os

def find_large_files(directory, size_limit_mb=100):
    """
    Recursively find all files larger than the specified size in megabytes.
    :param directory: The root directory to start searching from.
    :param size_limit_mb: The size limit in megabytes (default: 100MB).
    """
    # because os module finds the file size in bytes
    size_limit_bytes = size_limit_mb * 1024 * 1024
    large_files = []

    for root, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                if os.path.getsize(file_path) > size_limit_bytes:
                    large_files.append(file_path)
            except OSError as e:
                print(f"Error accessing {file_path} : {e}")

    return large_files
```

```python
# Create a script that rotates log files when they reach a certain size.
import shutil

def rotate_log_file(log_file_path, max_size_mb=5):
    """
    Rotates the log file when it exceeds the specified size limit.
    
    :param log_file_path: Path to the log file.
    :param max_size_mb: Maximum allowed log file size in megabytes before rotation.
    """
    max_size_bytes= max_size_mb * 1024**2   

    if os.path.exists(log_file_path) and os.path.getsize(log_file_path) >= max_size_bytes:
        old_log = f"{log_file_path}.0"
        new_log = f"{log_file_path}.1"

        if os.path.exists(old_log):
            os.rename(old_log, new_log)

        shutil.copy(log_file_path, f"{log_file_path}.1")
        open(log_file_path, "w").close() # clear the original log file
        print(f"log file rotated: {log_file_path}")
```

```python
# Merge two dictionaries recursively (deep merge)
def deep_merge(d1, d2):
    for k, v in d2.items(): # loop through all key-value pairs in d2
        # if the key also exists in d1
        # if v itself is a dictionary
        # if d1[k] is also a dictionary
        if k in d1 and isinstance(d1[k], dict) and isinstance(v, dict):
            # recursively merge nested dicts (recall this function but for the inner dicts)
            deep_merge(d1[k], v)
        else:
            d1[k] = v # otherwise, overwrite or add the value from d2 into d1
    return d1

a = {
    "app": {
        "version": "1.0",
        "settings": {
            "theme": "light"
        }
    }
}

b = {
    "app": {
        "settings": {
            "theme": "dark",
            "language": "en"
        }
    }
}

deep_merge(a, b)
```

```python
# Create a function that converts a string representing file size (e.g., "15MB", "2GB") into bytes.
import re

def filesize_to_bytes(size_str):
    size_units = {"B": 1, "KB": 1024, "MB": 1024**2, "GB": 1024**3, "TB": 1024**4}
    match = re.match(r"(\d+(\.\d+)?)\s*(B|KB|MB|GB|TB)", size_str, re.IGNORECASE)

    if not match:
        raise ValueError("Invalid file format")
    
    size, _, unit = match.groups()
    return int(float(size) * size_units[unit.upper()])

filesize_to_bytes("14.5GB")
```

```python
# Return SHA256 hash of a file.
import hashlib

def hash_file(file_path):
    # Initializes a new SHA-256 hash object from the hashlib module.
    h = hashlib.sha256()
    # open file in binary mode, Binary mode is important when hashing files — avoids issues with encoding differences
    with open(file_path, 'rb') as f:
        # read the file in 4096 bytes chunck
        # Using iter(lambda: f.read(4096), b"") keeps reading until f.read() returns an empty byte string (b""), meaning end of file (EOF)
        for chunck in iter(lambda: f.read(4096), b""):
            # Feeds each chunk into the hash function to update the hash state.
            h.update(chunck)
    # Returns the final hash value as a hexadecimal string
    return h.hexdigest()
```

```python
# Use boto3 to calculate total size of objects in an S3 bucket
import boto3

def get_bucket_size(bucket_name):
    # Initializes a Boto3 S3 client.
    s3 = boto3.client('s3')
    # S3 buckets may contain thousands or millions of files, To handle this, AWS paginates responses — it returns a limited number of objects per call
    # automatically fetch all the pages of objects in the bucket
    paginator = s3.get_paginator('list_objects_v2')
    total = 0
    # Iterates over each "page" of objects in the bucket
    for page in paginator.paginate(Bucket=bucket_name):
        # If the page has no 'Contents' key (i.e., the bucket is empty), .get(..., []) ensures the loop just skips
        for obj in page.get('Contents', []):
            # Access its 'Size' key (which is in bytes)
            total += obj['Size']
    return total
```

```python
# Create a function that extracts and counts HTTP status codes from a web server log file.
from collections import defaultdict

def count_http_status_codes(log_file_path):
    """
    Parses a web server log file to count the occurrences of each HTTP status code.
    Args:
        log_file_path (str): The path to the log file.
    Returns:
        dict: A dictionary with status codes as keys and their respective counts as values.
    """
    # Regular expression pattern to match the HTTP status code in a typical log entry : "GET /index.html HTTP/1.1" 200 512
    # \" – matches a double quote (") character. The backslash escapes the quote.
    # \d{3} – matches exactly 3 digits (e.g., 200, 404)
    # (\d{3}) – matches and captures another 3-digit number (in parentheses = capture group)
    log_pattern = re.compile(r'\" \d{3} (\d{3}) ')

    # Dictionary to hold the count of each status code
    status_counts = defaultdict(int)

    try:
        with open(log_file_path, "r") as file:
            for line in file:
                match = log_pattern.search(line)
                if match:
                    status_code = match.group(1)
                    status_counts[status_code] += 1
    except FileNotFoundError:
        print(f"Error not found the file at {log_file_path}")
    # general exception, anything else:
    except Exception as e:
        print(f"Error occurred: {e}")
    return dict(status_counts)
```

```python
# Write a script that watches a log file and sends an alert when a specific error pattern appears.
def watch_log_files(log_file_path, error_pattern, alert_callback, interval=10):
    """
    Watches a log file for a specific error pattern and triggers an alert callback when detected.
    
    :param log_file_path: Path to the log file.
    :param error_pattern: Regular expression pattern to match errors.
    :param alert_callback: Function to call when an error is found.
    :param interval: Time interval (in seconds) between file checks.
    """
    # example error pattern => r"ERROR|CRITICAL|FAILURE"
    error_regex = re.compile(error_pattern)

    with open(log_file_path, "r") as f:
        # f.seek(offset, whence) is a method used to move the cursor (aka file pointer) in a file object
        # move 0 bytes
        # 2: is the value of os.SEEK_END, which means "from the end of the file"
        f.seek(0, 2) # move to the end of the file

        while True:
            line = f.readline()
            if not line:
                time.sleep(interval)
                continue
            if error_regex.search(line):
                alert_callback(line)

def send_alert(message):
    print(f"ALERT {message} goes to correct api address")
```

```python
# Bubble sort 
# Bubble Sort Algorithm => Repeatedly compare adjacent elements, Swap them if they are in the wrong order, Largest elements “bubble up” to the end of the list
# Time Complexity: Worst-case: O(n²), Best-case (already sorted): O(n) with optimization
# Space Complexity: O(1) (in-place sorting)
# Stable Sort: Yes (does not change the relative order of equal elements)\

def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already sorted
        for j in range(0, n - i - 1):
            # swap if the element found is greater than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

arr = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(arr)
print("Sorted array:", arr)
```

```python
# Given a string, count how many of a certain letter appears
text = "hello world"
letter = "l"

count = text.count(letter)
print(f"{letter} : {count} times")

# another approach
text = "my friend marry"
letter = "m"
count = 0

for char in text:
    if char == letter:
        count += 1

print(f"{letter} : {count} times")
```

```python
# Given a list of chars that make a sentence, how would you identify the number of words (words are separated by ' ')
from collections import Counter

def count_words(sentence):
    # split sentence by spaces(its default) into a list
    words = sentence.split()
    word_freq = Counter(words)

    return len(words), word_freq

sentence = "hello world hello python python python"
total_words, frequencies = count_words(sentence)

print("Total words:", total_words)
print("Word frequencies:", frequencies)

for word, count in frequencies.items():
    print(f"{word} : {count}")
```

```python
# if you were given an unsorted list and you were looking for a specific element, how would you go about finding it?
''' 
If you have an unsorted list and want to find a specific element, the most straightforward approach is a linear search.
Linear Search (Sequential Search)
How it works:

Start from the first element of the list.
Check each element one by one.
If the element matches the one you’re looking for, return its position.
If you reach the end without a match, the element is not in the list.
'''
def linear_search(lst, target):
    for i, v in enumerate(lst):
        if v == target:
            return i # index where target is found
    return -1 # target not found

lst = [5, 3, 8, 2, 9]
target = 8
print(linear_search(lst, target)) 

# other approach, sort it first (with quick sort) then use binary search
```

*** What is Python?

Python is a programming language with objects, modules, threads, exceptions, and automatic memory management. The benefits of pythons are that it is simple and easy, portable, 
extensible, build-in data structure, and it is open-source.

*** What is PEP8?
PEP 8 is a coding convention, a set of recommendation, about how to write your Python code more readable.

*** What is pickling and unpickling?
Pickle module accepts any Python object and converts it into a "string representation" and dumps it into a file by using dump function. 
This process is called pickling. While the process of retrieving original Python objects from the stored string representation is called unpickling.

*** How is Python interpreted?
Python language is an interpreted language. Python program runs directly from the source code. It converts the source code that is written by the programmer into an intermediate language, which is again translated into machine language that has to be executed.

*** How is memory managed in python?
Python memory is managed by Python "private heap space". All Python objects and data structures are located in a private heap. The programmer does not have an access to this 
private heap, and the interpreter takes care of this Python private heap. The allocation of Python heap space for Python objects is done by the Python memory manager.
The core API gives access to some tools for the programmer to code. Python also has an inbuilt garbage collector, which recycles all the unused memory and frees 
the memory and makes it available to the heap space.

*** How are arguments passed by value or by reference?
Everything in Python is an object, and all variables hold references to the objects. The reference values are according to the functions. Therefore, you cannot change the 
value of the references. However, you can change the objects if it is mutable.

*** Explain namespace in Python
In Python, every name introduced has a place where it lives and can be looked for. This is known as a namespace. It is like a box where a 
variable name is mapped to the object placed. Whenever the variable is searched out, this box will be searched to get the corresponding object.
local => enclosed => global => built-in

*** Python Scopes?
local scope => refers to the local objects available in the current function.
global scope => refers to the objects available throught the code execution since their inception.
module-level scope => refers to the global objects of the current module accessible in the program.
outermost scope => refers to all the built-in names callable in the program. The objects in this scope are searched last to find the name referenced.

*** In Python what are iterators?
In Python, iterators are used to iterate a group of elements, containers like a list.

*** What is docstring in Python?
A Python documentation string is known as docstring, it is a way of documenting Python functions, modules, and classes.

*** What is negative index in Python?
Python sequences can be index in positive and negative numbers. For positive index, 0 is the first index, 1 is the second index, 
and so forth. For the negative index, (-1) is the last index, and (-2) is the second last index, and so forth.

*** What is module and package in Python?
In Python, module is the way to structure a program. Each Python program "file" is a module, which imports other modules like objects and attributes.
The "folder" of Python program is a package of modules. A package can have modules or subfolders.
Modules, in general, are simply Python files with a .py extension and can have a set of functions, classes or variables defined and 
implemented. They can be imported and initialized once using the import statement. If partial functionality is needed, import the 
requisite classes or functions using "from foo import bar".
Packages allow for hierarchial structuring of the module namespace using dot notation. As, modules help avoid clashes between global 
variable names, in a similary manner, packages help avoid clashes between module names.

*** What is self in Python?
Self is a keyword in Python used to define an instance or an object of a class. In Python, it is explicity used as the first parameter, 
unlike in Java where it is optional. It helps in disinguishing between the methods and attributes of a class from its local variables.

*** Python Break, Continue, Pass?
break => The break statement terminates the loop immediately and the control flows to the statement after the body of the loop.
continue => The continue statement terminates the current iteration of the statement, skips the rest of the code in the current iteration and the control flows to the next iteration of the loop.
pass => pass keyword in Python is generally used to fill-up empty block

*** help() and dir() ?
help() => help function in Python is used to display the documentation of modules, classes, functions, keywords, etc. If no parameter is 
passed to the help() function, then an interactive help utility is launched on the console.
dir => dir() function tries to return a valid list of attributes and methods of the object it is called upon. It behaves differently 
with different objects, as it aims to produce the most relevant data, rather than the complete information.

*** .py vs .pyc ?
".py" files contain the source code of a program. Whereas, ".pyc" file contains the bytecode of your program. We get bytecode after compilation of .py file (source code). 
.pyc files are not created for all the files that you run. It is only created for the files that you import.

Is Java pass-by-value or by reference?
Java handles objects and primitives differently.
Java is always pass-by-value, Pass-by-value means Java passes a copy of the variable to the method, not the original variable itself. 
For primitives, the method gets a copy of the value.
For objects, the method gets a copy of the reference to the object.
