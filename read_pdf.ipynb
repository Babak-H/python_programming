```python
import PyPDF2

# this will read the text(evrything not just table) from third page, each cell of table becomes one line
with open('./Files/Food.pdf', 'rb') as pdf_file:
    read_file = PyPDF2.PdfFileReader(pdf_file)
    number_of_pages = read_file.getNumPages() # get all the pages
    page = read_file.getPage(2) # third page : 0 1 2
    page_content = page.extractText()
    print(page_content)
```

```python
import numpy

# turn it into a list
table_list = page_content.split('\n') 

# this function will turn our list to multiple 4 element lists
# we have 128 element in table_list -> l = 32 different 4-element sublists
l = numpy.array_split(table_list, len(table_list)/4)

# loop through first 5 rows
for i in range(0, 5):
    print(l[i])
```

# Read PDF Text

```python
import PyPDF2 as p2

with open("./Files/sample.pdf", 'rb') as file:
    read_file = p2.PdfFileReader(file)
    num_pages = read_file.getNumPages()
    print(num_pages)
    print("")
    text_data = read_file.getPage(0).extractText()
    print(text_data)
    print("")
    print(text_data[3]) # read third character from text
    
    print(read_file.getDocumentInfo()) # information about pdf file
```

```python
# Extract the entine doc

with open("./Files/sample.pdf", 'rb') as file:
    read_file = p2.PdfFileReader(file)
    
    i = 0
    while i < read_file.getNumPages():

        pageText = read_file.getPage(i).extractText()
        print(pageText, "\n")
        i = i + 1
```
